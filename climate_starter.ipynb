# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, desc
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()

# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# print measurent table first row
measurement_first_row = session.query(Measurement).first()
measurement_first_row.__dict__
```

```python
# print station table first row
station_first_row = session.query(Station).first()
station_first_row.__dict__
```

# Exploratory Climate Analysis

## Design a query to retrieve the last 12 months of precipitation data and plot the result

```python
# Calculate the date 1 year ago from the last data point in the database
last_date = session.query(func.max(Measurement.date)).scalar()
print(f'last_date: {last_date}')

last_date_minus_12m = (dt.datetime.strptime(last_date, '%Y-%m-%d') - dt.timedelta(days=365)).date()
print(f'Date 1 year ago from last_date: {last_date_minus_12m}')
```

```python
# Perform a query to retrieve the date and precipitation scores
last_12month_prcp = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date > last_date_minus_12m).all()
last_12month_prcp
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
last_12month_prcp_df = pd.DataFrame(last_12month_prcp)
last_12month_prcp_df.set_index('date', inplace=True)
last_12month_prcp_df.dropna(inplace=True)
last_12month_prcp_df
```

```python
# Sort the dataframe by date
last_12month_prcp_df.sort_index(inplace=True)
last_12month_prcp_df
```

```python
# check precipitation min and max
min = last_12month_prcp_df['prcp'].min()
max = last_12month_prcp_df['prcp'].max()

min, max
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
last_12month_prcp_df.plot(kind='line', title="Precipitaion by Date", figsize=(15,9))  # default kind is 'line'
plt.ylabel("Inches")
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
last_12month_prcp_df['prcp'].describe()
```

```python
# Design a query to show how many stations are available in this dataset?
station_count = session.query(Station.id).count()
station_count
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.

st_count = func.count(Measurement.station)
stations_active = session.query(Measurement.station, st_count).group_by(Measurement.station).order_by(st_count.desc()).all()
stations_active
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?
most_active_station = stations_active[0][0]
print(most_active_station)
#min = session.query(func.min(Measurement.tobs)).filter_by(station = most_active_station).all()
min = session.query(func.min(Measurement.tobs)).filter_by(station = most_active_station).scalar()
max = session.query(func.max(Measurement.tobs)).filter_by(station = most_active_station).scalar()
mean = session.query(func.avg(Measurement.tobs)).filter_by(station = most_active_station).scalar()

min, max, mean
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
tobs_last_12m = session.query(Measurement.date, Measurement.tobs)\
                  .filter_by(station = most_active_station)\
                  .filter(Measurement.date > last_date_minus_12m).all()

tobs_last_df = pd.DataFrame(tobs_last_12m)
tobs_last_df.set_index('date', inplace=True)
tobs_last_df.dropna(inplace=True)
tobs_last_df
```

```python
# plot a historgram of Temperature
tobs_last_df.plot(kind='hist', title="Temperature Histogram", figsize=(14,9), bins=12)
plt.xlabel("Temperature")
```

## Temperature Analysis II  - Daily Temperature Average

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
start_date = '2017-06-05'
end_date = '2017-06-18'
tmin, tavg, tmax = calc_temps(start_date, end_date)[0]
print(tmin, tavg, tmax)
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
plt.figure(figsize=(3,7))
plt.bar(x="", height=tavg, yerr=(tmax-tmin), width = 3, color = 'c')
plt.ylabel("Temp (F)")
plt.title("Trip Avg Temp")
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation
sel = [Measurement.station, func.sum(Measurement.prcp).label("Total Rain Fall"),\
       Station.name, Station.latitude, Station.longitude, Station.elevation]
total_rainfall = session.query(*sel)\
            .filter(Measurement.station == Station.station)\
            .filter(Measurement.date >= start_date).filter(Measurement.date <= end_date)\
            .group_by(Measurement.station)\
            .order_by(desc("Total Rain Fall")).all()

total_rainfall_df = pd.DataFrame(total_rainfall)
total_rainfall_df
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("06-15")
```

```python
# calculate the daily normals for your trip
# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip
start_date = '2017-06-05'
end_date = '2017-06-18'

# Use the start and end date to create a range of dates
date_list = [dt.datetime.strptime(start_date, '%Y-%m-%d').date()  + dt.timedelta(days=x) for x in range(18-5+1)]
date_list

#Stip off the year and save a list of %m-%d strings
date_list_md = []
for d in date_list:
    date_list_md.append(f"{d.month:02}-{d.day:02}")
date_list_md

#Loop through the list of %m-%d strings and calculate the normals for each date
normals = []
for d in date_list_md:
    normals.append(daily_normals(d))

normals
```

```python
# Load the previous query results into a Pandas DataFrame and add the `trip_dates` range as the `date` index
normals_prep = []

for i in normals:
    d = {}
    d['tmin'], d['tavg'], d['tmax'] = i[0]
    normals_prep.append(d)
        
normals_df = pd.DataFrame(normals_prep)
normals_df['date'] = date_list
normals_df.set_index('date', inplace=True)
normals_df
```

```python
# Plot the daily normals as an area plot with `stacked=False`
normals_df.plot(kind='area', stacked=False, title="Temperature Daily Normals", figsize=(15,8))
plt.ylabel("Temperature")
```
